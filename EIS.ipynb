Here is the plot of Impedance vs. Frequency for the RLC circuit:

The x-axis represents the frequency in Hertz (Hz) on a logarithmic scale.
The y-axis represents the magnitude of the impedance in Ohms, also on a logarithmic scale.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
R = 100  # ohms
L = 1e-3  # henrys
C = 1e-6  # farads
frequencies = np.logspace(1, 6, 1000)
omega = 2 * np.pi * frequencies

# Impedance calculation
Z_R = R
Z_L = 1j * omega * L
Z_C = 1 / (1j * omega * C)
Z_RLC = Z_R + Z_L + Z_C

# Bode plot (Impedance vs Frequency)
plt.figure(figsize=(10, 6))
plt.loglog(frequencies, np.abs(Z_RLC), label='Magnitude')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude of Impedance (Ohms)')
plt.title('Impedance vs Frequency for RLC Circuit')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the parallel RC and series R2 circuit
frequencies = np.logspace(1, 6, 1000)
omega = 2 * np.pi * frequencies
R1 = 100  # ohms
R2 = 100  # ohms
C = 1e-6  # farads

# Impedance calculation for R1 and C in parallel
Z_R1 = R1 * np.ones_like(frequencies)
Z_C = 1 / (1j * omega * C)
Z_parallel_RC = 1 / (1 / Z_R1 + 1 / Z_C)

# Total impedance for (R1||C) in series with R2
Z_total = Z_parallel_RC + R2

# Plotting
plt.figure(figsize=(10, 6))
plt.loglog(frequencies, np.abs(Z_total), label='(R1||C) + R2')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Impedance (Ohms)')
plt.title('Impedance vs Frequency for (R1||C) + R2 Circuit')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```

```python
# Phase calculation for (R1||C) in series with R2
phase_total = np.angle(Z_total, deg=True)

# Plotting
plt.figure(figsize=(10, 6))
plt.semilogx(frequencies, phase_total, label='(R1||C) + R2')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (Degrees)')
plt.title('Phase vs Frequency for (R1||C) + R2 Circuit')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```

### Nyquist plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the parallel RC and series R2 circuit
frequencies = np.logspace(1, 6, 1000)
omega = 2 * np.pi * frequencies
R1 = 200  # ohms
R2 = 20  # ohms
C = 1e-6  # farads

# Impedance calculation for R1 and C in parallel
Z_R1 = R1 * np.ones_like(frequencies)
Z_C = 1 / (1j * omega * C)
Z_parallel_RC = 1 / (1 / Z_R1 + 1 / Z_C)

# Total impedance for (R1||C) in series with R2
Z_total = Z_parallel_RC + R2

# Extracting real and imaginary parts
Z_real_total = Z_total.real
Z_imag_total = Z_total.imag

# Plotting Z(imaginary) vs Z(real)
plt.figure(figsize=(10, 8))

plt.plot(Z_real_total, -Z_imag_total, label='Total Impedance Trajectory')
plt.title('Imaginary vs Real Part of Total Impedance (Z)')
plt.xlabel('Z (Real Part)')
plt.ylabel('- Z (Imaginary Part)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.show()
```

## add Warburg element

```python
# Adding Warburg element
# Constants for Warburg element
w = 1000
pi_over_4 = np.pi / 4

# Warburg impedance calculations
Z_Warburg_magnitude = w / np.sqrt(frequencies)
Z_Warburg_phase = -pi_over_4 * np.ones_like(frequencies)
Z_Warburg = Z_Warburg_magnitude * (np.cos(Z_Warburg_phase) + 1j * np.sin(Z_Warburg_phase))

# Total impedance including Warburg element
Z_total_with_Warburg = Z_total + Z_Warburg

# Extracting real and imaginary parts
Z_real_total = Z_total.real
Z_imag_total = Z_total.imag

Z_real_total_with_Warburg = Z_total_with_Warburg.real
Z_imag_total_with_Warburg = Z_total_with_Warburg.imag

# Plotting Z(imaginary) vs Z(real)
plt.figure(figsize=(10, 8))

plt.plot(Z_real_total, -Z_imag_total, label='Total Impedance Trajectory')
plt.plot(Z_real_total_with_Warburg, -Z_imag_total_with_Warburg, label='Total Impedance with Warburg', linestyle='--')
plt.title('Imaginary vs Real Part of Total Impedance (Z)')
plt.xlabel('Z (Real Part)')
plt.ylabel('- Z (Imaginary Part)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.show()
```

```python
# Recalculate Z_Warburg correctly
# Warburg impedance has a real and imaginary part both influenced by the sqrt(frequency)
Z_Warburg = Z_Warburg_magnitude * (np.cos(Z_Warburg_phase) + 1j * np.sin(Z_Warburg_phase))

# Total impedance including Warburg element
Z_total_with_Warburg = Z_parallel_RC + R2 + Z_Warburg

# Extracting real and imaginary parts
Z_real_total_with_Warburg = Z_total_with_Warburg.real
Z_imag_total_with_Warburg = Z_total_with_Warburg.imag

# Plotting Z(imaginary) vs Z(real)
plt.figure(figsize=(10, 8))

plt.plot(Z_real_total, -Z_imag_total, label='Total Impedance Trajectory')
plt.plot(Z_real_total_with_Warburg, -Z_imag_total_with_Warburg, label='Total Impedance with Warburg', linestyle='--')
plt.title('Imaginary vs Real Part of Total Impedance (Z)')
plt.xlabel('Z (Real Part)')
plt.ylabel('- Z (Imaginary Part)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.show()
```
